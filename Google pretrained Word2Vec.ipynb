```python
import gensim
from gensim.models import Word2Vec, KeyedVectors
import pandas as pd
import nltk
```

```python
model = KeyedVectors.load_word2vec_format('GoogleNews-vectors-negative300.bin', binary=True, limit=100000)
```

```python
model.wv.most_similar('hero')
```

```python
model.wv['man']
```

```python
vec = model['Dhoni'] - model['cricket'] + model['wicketkeeper']
model.most_similar([vec])
```

```python
vec2 = model['USA'] - model['Washington'] + model['Delhi']
model.most_similar([vec2])
                                                
```

```python
score = model.wv.similarity('man', 'boy')
score
```
